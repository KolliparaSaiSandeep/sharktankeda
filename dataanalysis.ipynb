## How does the distribution of 'pitcher_ask_amount' vary across different episodes or pitch numbers?


```python
groupedbypitchnumber=df.groupby('pitch_number')
pitcher_ask_amount = groupedbypitchnumber['pitcher_ask_amount'].sum()
pitcher_ask_amount
```

```python
import matplotlib.pyplot as plt

# Group by 'episode_number' and sum up 'pitcher_ask_amount' for each episode
episode_ask_amount = df.groupby('episode_number')['pitcher_ask_amount'].sum()
print(episode_ask_amount)
# Plot combined ask amounts for each episode
plt.figure(figsize=(10, 6))
episode_ask_amount.plot(kind='bar', color='skyblue')
plt.title('Combined Pitcher Ask Amount for Each Episode')
plt.xlabel('Episode Number')
plt.ylabel('Combined Pitcher Ask Amount')
for i, value in enumerate(episode_ask_amount):
    plt.text(i, value, f'{value:.2f}', ha='center', va='bottom', rotation=90)
plt.tight_layout()
plt.show()
```

# What is the distribution of 'ask_equity' among the pitchers? Are there any outliers?


```python
plt.figure(figsize=(10, 6))
sns.histplot(df['ask_equity'], bins=20, kde=True)
plt.title('Distribution of Ask Equity')
plt.xlabel('Ask Equity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Identify Outliers
mean = df['ask_equity'].mean()
std_dev = df['ask_equity'].std()
outlier_threshold = mean + 3 * std_dev  # Example: consider values beyond 3 standard deviations from the mean as outliers
outliers = df[df['ask_equity'] > outlier_threshold]
print("\nNumber of outliers:", len(outliers))
print("Outlier values:\n", outliers)#this means these 2 asks are unusual
```

# How does the 'deal_amount' compare to the pitcher_ask_amount'? Are there any trends or patterns?

```python
import matplotlib.pyplot as plt

# Create a scatter plot to compare 'pitcher_ask_amount' and 'deal_amount'
plt.figure(figsize=(10, 6))
plt.scatter(df['pitcher_ask_amount'], df['deal_amount'], alpha=0.5)
plt.title('Comparison of Pitcher Ask Amount and Deal Amount')
plt.xlabel('Pitcher Ask Amount')
plt.ylabel('Deal Amount')
plt.show()

# Calculate summary statistics
correlation_coefficient = df['pitcher_ask_amount'].corr(df['deal_amount'])
print("Correlation Coefficient:", correlation_coefficient)
```

# What is the distribution of 'deal_equity' among the pitchers who successfully strike a deal?

```python
successful_deals_df = df[df['deal']==1]
deal_equity_counts = successful_deals_df['deal_equity'].value_counts()
# Plot the bar plot
plt.figure(figsize=(10, 6))
deal_equity_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Deal Equity for Successful Deals')
plt.xlabel('Deal Equity')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# Differences in ask_equity vs deal_equity in successful deals


```python
import matplotlib.pyplot as plt
import numpy as np

# Filter successful deals
grouped_df = df[df['deal'] == 1]

# Extract unique company names
company_names = grouped_df['brand_name'].unique()

# Initialize lists to store ask and deal equity values for each company
ask_equity_values = []
deal_equity_values = []

# Iterate through each company and calculate total ask and deal equity
for company in company_names:
    ask_equity_values.append(grouped_df[grouped_df['brand_name'] == company]['ask_equity'].sum())
    deal_equity_values.append(grouped_df[grouped_df['brand_name'] == company]['deal_equity'].sum())

# Set the width of the bars
bar_width = 0.35

# Create the figure and axes objects
fig, ax = plt.subplots(figsize=(12, 6))

# Plot ask equity bars
ax.bar(np.arange(len(company_names)) - bar_width/2, ask_equity_values, bar_width, label='Ask Equity')

# Plot deal equity bars
ax.bar(np.arange(len(company_names)) + bar_width/2, deal_equity_values, bar_width, label='Deal Equity')

# Add labels, title, and legend
ax.set_xlabel('Company Name')
ax.set_ylabel('Equity')
ax.set_title('Ask Equity vs. Deal Equity by Company in Successful Deals')
ax.set_xticks(np.arange(len(company_names)))  # Setting x-ticks
ax.set_xticklabels(company_names, rotation=90, ha='right')  # Setting x-tick labels
ax.legend()

# Show plot
plt.tight_layout()
plt.show()
```
